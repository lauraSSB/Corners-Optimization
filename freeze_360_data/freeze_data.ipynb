```python
username = os.getenv("SB_USERNAME")
password = os.getenv("SB_PASSWORD")
```

```python
creds = {
"user": username,
"passwd": password,
}
```

```python
df = pd.read_csv('/Users/arturoreza/Documents/isac_2025/matches_data/all_matches.csv')
```

```python
season_2024_2025 = df.loc[df['season'] == '2024/2025']['match_id'].unique().tolist()
season_2023_2024 = df.loc[df['season'] == '2023/2024']['match_id'].unique().tolist()
season_2022_2023 = df.loc[df['season'] == '2022/2023']['match_id'].unique().tolist()
season_2021_2022 = df.loc[df['season'] == '2021/2022']['match_id'].unique().tolist()
```

```python
# Make a folder to store the files
os.makedirs("season_2023_2024_freeze_data", exist_ok=True)

for match_id in season_2023_2024:
    url = f'https://data.statsbombservices.com/api/v2/360-frames/{match_id}'
    response = requests.get(url, auth=HTTPBasicAuth(username, password))
    
    if response.status_code == 200:
        frames = response.json()
        # ✅ Save this match's frames to its own JSON file
        with open(f"season_2023_2024_freeze_data/{match_id}.json", "w") as f:
            json.dump(frames, f)
        print(f"Saved 360 data for match {match_id}")
    else:
        print(f"Failed for match {match_id}: {response.status_code}")
```
